```python
cities_of_interest = ['Anchorage', 'Boise', 'Chicago', 'Denver', 'Detroit', 'Honolulu', 'Houston', 'Miami',  'Minneapolis',
                      'Oklahoma City', 'Nashville', 'New York', 'Phoenix', 'Portland ME', 'Portland OR', 'Salt Lake City',
                      'San Diego','San Francisco', 'Seattle','Washington DC']
```

```python
kaggle_cities = k_df['City'].unique()
```

```python
missing = []
for city in cities_of_interest:
    if city not in kaggle_cities:
        missing.append(city)
print(missing)
```

```python
from geopy.distance import geodesic
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="city_distance_calculator")

def get_coordinates(city_name):
    location = geolocator.geocode(city_name)
    return (location.latitude, location.longitude)

def calculate_distance(city1, city2):
    coords_1 = get_coordinates(city1)
    coords_2 = get_coordinates(city2)
    return geodesic(coords_1, coords_2).kilometers
```

```python
distances = {}
for city in cities_of_interest:
    distances[city] = {}  # Initialize an inner dictionary for each city in cities of interest
    for c in kaggle_cities:
        # Calculate the distance between city and c
        distance = calculate_distance(city, c)
        print(city, c, distance)
        # Store the distance in the inner dictionary
        distances[city][c] = distance
```

```python
distances
```

```python
nearest_cities = {}

for city, neighbors in distances.items():
    # Sort the neighboring cities by distance and select the 5 nearest
    sorted_neighbors = sorted(neighbors.items(), key=lambda item: item[1])
    nearest_cities[city] = dict(sorted_neighbors[:5])  # Take the top 5

# nearest_cities now contains the 5 nearest cities for each key city
print(nearest_cities)
```
