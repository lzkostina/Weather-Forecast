```python
import pandas as pd
import numpy as np

# Combine Year, Month, and Day into a single column for datetime conversion (no need for 'Time')

hourly_data['Datetime'] = pd.to_datetime(hourly_data[['Year', 'Month', 'Day']].astype(str).agg('-'.join, axis=1) + ' ' + hourly_data['Time'])

# Filter the hourly data for the date range from 2023-11-29 to the last available date
start_date = pd.to_datetime('2023-11-30')
end_date = hourly_data['Datetime'].max()  # Last available date in the dataset
filtered_hourly_data = hourly_data[(hourly_data['Datetime'] >= start_date) & (hourly_data['Datetime'] <= end_date)]

# Resample the hourly data to daily aggregates
daily_data = filtered_hourly_data.resample('D', on='Datetime').agg({
    'Temperature (F)': 'mean',  # Average temperature
    'Feels Like (F)': 'mean',
    'Temp Min (F)': 'min',  # Minimum temperature of the day
    'Temp Max (F)': 'max',  # Maximum temperature of the day
    'Pressure': 'mean',  # Average pressure
    'Humidity': 'mean',  # Average humidity
    'Wind Speed': 'mean',  # Average wind speed
    'Wind Deg': 'mean',  # Average wind direction
    'Wind Gust': 'max',  # Maximum wind gust
    'Clouds All': 'mean',  # Average cloud coverage
    'Weather Description': lambda x: x.mode()[0]  # Most frequent weather description
}).reset_index()

# Extract YEAR, MONTH, and DAY columns from 'Datetime' in daily_data
daily_data['YEAR'] = daily_data['Datetime'].dt.year
daily_data['MONTH'] = daily_data['Datetime'].dt.month
daily_data['DAY'] = daily_data['Datetime'].dt.day

# Merge the data with the daily aggregated data based on YEAR, MONTH, and DAY
merged_data = pd.merge(data, daily_data[['YEAR', 'MONTH', 'DAY', 'Temperature (F)', 'Feels Like (F)', 
                                          'Temp Min (F)', 'Temp Max (F)', 'Pressure', 'Humidity', 'Wind Speed',
                                          'Wind Deg', 'Wind Gust', 'Clouds All', 'Weather Description']], 
                       on=['YEAR', 'MONTH', 'DAY'], how='left')

# Check if there are any missing values (NaNs) in the merged dataset
print(merged_data.isna().sum())  # Check for missing data after merging

# Fill missing data by forward filling (or using another method such as backward fill, mean, etc.)
merged_data.fillna(method='ffill', inplace=True)  # Forward fill missing values

# Optionally, save the filled data to a new CSV
merged_data.to_csv('merged_filled_data.csv', index=False)

print("Missing data has been filled and saved to 'merged_filled_data.csv'.")
```
